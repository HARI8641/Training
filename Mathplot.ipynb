# Library Required

```python
import pandas as pd
```

```python
df = pd.DataFrame({'x': [1, 2, 3, 4], 'y': [1, 4, 9, 16]})
```

```python
df.plot(x='x', y='y', kind='line')
```

```python
df.plot(x='x', y='y', kind='bar', legend=True, title='Bar Plot')
```

```python
df.plot(x='x', y='y', kind='scatter')
```

```python
df.plot(x='x', y='y', kind='box')
```

```python
df.plot(x='x', y='y', kind='hist')
```
